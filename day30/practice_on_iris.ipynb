```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import FunctionTransformer
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
df = pd.DataFrame(data=X, columns=iris.feature_names)
```

```python
df.head()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.histplot(df['petal length (cm)'], kde=True)
plt.title('Petal Length before')

plt.subplot(122)
stats.probplot(df['petal length (cm)'], dist="norm", plot=plt)
plt.title('Petal Length before')

plt.show()
```

```python
log_transformer = FunctionTransformer(np.log1p) # log(1 + x)
df_log = pd.DataFrame(log_transformer.fit_transform(df), columns=df.columns)
```

```python
df_log.head()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.histplot(df_log['petal length (cm)'], kde=True)
plt.title('Petal Length after (log)')

plt.subplot(122)
stats.probplot(df_log['petal length (cm)'], dist="norm", plot=plt)
plt.title('Petal Length after (log)')

plt.show()
```

```python
reciprocal_transformer = FunctionTransformer(lambda x: 1 / (x + 1e-5))
df_recip = pd.DataFrame(reciprocal_transformer.fit_transform(df), columns=df.columns)
```

```python
df_recip.head()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.histplot(df_recip['petal length (cm)'], kde=True)
plt.title('Petal Length after (reciprocal)')

plt.subplot(122)
stats.probplot(df_recip['petal length (cm)'], dist="norm", plot=plt)
plt.title('Petal Length after (reciprocal)')

plt.show()
```

```python
sqrt_transformer = FunctionTransformer(np.sqrt)
df_sqrt = pd.DataFrame(sqrt_transformer.fit_transform(df), columns=df.columns)
```

```python
df_sqrt.head()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.histplot(df_sqrt['petal length (cm)'], kde=True)
plt.title('Petal Length after (sqrt)')

plt.subplot(122)
stats.probplot(df_sqrt['petal length (cm)'], dist="norm", plot=plt)
plt.title('Petal Length after (sqrt)')

plt.show()
```

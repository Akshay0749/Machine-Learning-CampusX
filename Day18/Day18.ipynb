### Dataframe for 1 Webpage

```python
name = []
rating = []
reviews = []
salary = []
interviews = []
jobs = []
benefits = []


for i in companyInfo:
    name.append(i.find('h2').text.strip()) #find because only 1 company in h2 tag
    rating.append(i.find('div', class_='rating_text rating_text--md').text.strip())

    actions = i.find_all('span', class_='companyCardWrapper__ActionCount')
    reviews.append(actions[0].text.strip())
    salary.append(actions[1].text.strip())
    interviews.append(actions[2].text.strip())
    jobs.append(actions[3].text.strip())
    benefits.append(actions[4].text.strip())    

df = pd.DataFrame({'name':name, 'rating':rating, 'reviews':reviews, 'salary':salary, 'interviews':interviews, 'jobs':jobs, 'benefits':benefits})
```

### Dataframe for top 250 companies in Pune

```python
final = pd.DataFrame()
for j in range(1, 101):
    time.sleep(3)
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(f"https://www.ambitionbox.com/companies-in-pune?page={j}",headers=headers).text
    soup = BeautifulSoup(webpage, 'lxml')
    company = soup.find_all('div', class_='companyCardWrapper')
    name = []
    rating = []
    reviews = []
    salary = []
    interviews = []
    jobs = []
    benefits = []

    for i in company:
        try:
            name.append(i.find('h2').text.strip()) #find because only 1 company in h2 tag
        except:
            name.append(np.nan)
            
        try:
            rating.append(i.find('div', class_='rating_text rating_text--md').text.strip())
        except:
            rating.append(np.nan)
            
        actions = i.find_all('span', class_='companyCardWrapper__ActionCount')
        
        try:
            reviews.append(actions[0].text.strip())
        except:
            reviews.append(np.nan)
            
        try:
            salary.append(actions[1].text.strip())
        except:
            salary.append(np.nan)
    
        try:
            interviews.append(actions[2].text.strip())
        except:
            interviews.append(np.nan)
    
        try:
            jobs.append(actions[3].text.strip())
        except:
            jobs.append(np.nan)
    
        try:
            benefits.append(actions[4].text.strip())
        except:
            benefits.append(np.nan)
        
    df = pd.DataFrame({'name':name, 'rating':rating, 'reviews':reviews, 'salary':salary, 'interviews':interviews, 'jobs':jobs, 'benefits':benefits})

    final = pd.concat([final, df], ignore_index=True)
```

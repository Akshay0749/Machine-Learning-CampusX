# Predict Using Pipelines

```python
import pickle
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
pipe = pickle.load(open('pipe.pkl', 'rb'))
```

```python
# Format: [Pclass, Sex, Age, SibSp, Parch, Fare, Embarked]
test_input = np.array([2, 'male', 31.0, 0, 0, 10.5, 'S'], dtype=object).reshape(1, 7)
```

```python
result = pipe.predict(test_input)
print("Survived!" if result[0] == 1 else "Did not survive")
```

```python
test_input1 = np.array([1, 'female', 29.0, 0, 0, 100.0, 'C'], dtype=object).reshape(1,7)
test_input2 = np.array([3, 'male', 20.0, 0, 0, 7.25, 'S'], dtype=object).reshape(1,7)
test_input3 = np.array([1, 'male', 55.0, 0, 0, 150.0, 'C'], dtype=object).reshape(1,7)
```

```python
result1 = pipe.predict(test_input1) #survived
result2 = pipe.predict(test_input2) #did not survived
result3 = pipe.predict(test_input3) #survived
print("Survived!" if result1[0] == 1 else "Did not survive")
print("Survived!" if result2[0] == 1 else "Did not survive")
print("Survived!" if result3[0] == 1 else "Did not survive")
```

```python
test_data = pd.read_csv('test.csv')
```

```python
X_test = test_data[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
```

```python
pred = pipe.predict(X_test)
```

```python
submission = pd.DataFrame({
    "PassengerId": test_data["PassengerId"],
    "Survived": pred
})
```

```python
submission.to_csv("submission.csv", index=False)
```

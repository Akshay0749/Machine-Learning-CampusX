FETCHING DATA FROM API

```python
import pandas as pd
import numpy as np
import requests
import time
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=67e8e9631af99a55c87632a64337a3bb&language=en-US&page=1')
```

```python
pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_count', 'vote_average']]
```

```python
df = pd.DataFrame()
```

```python
df
```

```python
for i in range(1, 535):
    try:
        response = requests.get(
            'https://api.themoviedb.org/3/movie/top_rated?api_key=67e8e9631af99a55c87632a64337a3bb&language=en-US&page={}'.format(i),
            timeout=10   # 🔥 prevents hanging connections
        )

        if response.status_code != 200:
            print("Skipping page:", i)
            continue

        temp_df = pd.DataFrame(response.json()['results'])[
            ['id', 'title', 'overview', 'release_date', 'popularity', 'vote_count', 'vote_average']
        ]

        df = pd.concat([df, temp_df], ignore_index=True)
        print("Done page:", i)

        time.sleep(0.45)

    except:
        print("Retrying page:", i)
        time.sleep(2)
```

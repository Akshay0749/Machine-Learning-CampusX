# EDA using Univariate Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
df.head()
```

## 1.Categorical Data

### a. Countplot

```python
sns.countplot(df['Embarked'])
# df['Survived'].value_counts().plot(kind='bar')
```

### b. PieChart

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f')
```

## 2. Numerical Data

### a. Histogram

```python
plt.hist(df['Age'], bins=10) #numerical data can be beahved as categorical data
```

### b. Distplot

```python
sns.displot(df['Age'], kde=True) #Distplot is removed use kde = True in displot --> used to plot kde(kernel density estimation) curve
```

### c. Boxplot

```python
sns.boxplot(df['Age'], orient='h')
```

```python
df['Age'].min()
```

```python
df['Age'].max()
```

```python
df['Age'].mean() #average age of passengers
```

```python
df['Age'].median() #middle value of Age column
```

```python
df['Age'].mode() #most frequent age
```

```python
df['Age'].skew()
```

## Survival by Passeneger Class

```python
survival_by_class = pd.crosstab(df['Pclass'], df['Survived'])
survival_by_class.plot(kind='bar')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Survival by Passenger Class')
plt.legend(['Did not survive', 'Survived'])
```
